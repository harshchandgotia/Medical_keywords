```python
import pandas as pd
df1 = pd.read_excel('Compiled(1).xlsx', sheet_name=1)
df2 = pd.read_excel('Compiled(1).xlsx', sheet_name=6)


df1.drop(df1.columns[[0,1,2,3,4,5,6,8,11,12, 14, 16, 17, 18]], axis=1, inplace=True)


df2.drop(df2.columns[[0,1,4,5,6,8,9,10]], axis=1, inplace=True)
df = pd.concat([df1, df2], axis = 1)
```

```python
df['Diagnosis'] = df['Diagnosis'].apply(lambda x:str(x).lower())
df['VOC/SCC'] = df['VOC/SCC'].apply(lambda x:str(x).lower())
df['Complaint'] = df['Complaint'].apply(lambda x:str(x).lower())
df['Investigation'] = df['Investigation'].apply(lambda x:str(x).lower())
df['Procedure'] = df['Procedure'].apply(lambda x:str(x).lower())
df['Special notes'] = df['Special notes'].apply(lambda x:str(x).lower())
df['Problem'] = df['Problem'].apply(lambda x:str(x).lower())
df['Examination'] = df['Examination'].apply(lambda x:str(x).lower())
```

```python
with open ("keywords.txt", "r") as f:
    text = f.readlines()

tokens = [line.strip() for line in text]
tokens = [line.lower() for line in tokens]
```

```python
token_freq = {k: 0 for k in tokens}
print(token_freq)
```

```python
for words in df['Diagnosis']:
    print(words)
```

```python
for words in df['Diagnosis']:
    for keyword in token_freq:
        if str(keyword) in str(words):
            token_freq[keyword] += 1


for words in df['VOC/SCC']:
    for keyword in token_freq:
        if str(keyword) in str(words):
            token_freq[keyword] += 1


for words in df['Complaint']:
    for keyword in token_freq:
        if str(keyword) in str(words):
            token_freq[keyword] += 1


for words in df['Investigation']:
    for keyword in token_freq:
        if str(keyword) in str(words):
            token_freq[keyword] += 1

for words in df['Procedure']:
    for keyword in token_freq:
        if str(keyword) in str(words):
            token_freq[keyword] += 1


for words in df['Special notes']:
    for keyword in token_freq:
        if str(keyword) in str(words):
            token_freq[keyword] += 1

for words in df['Problem']:
    for keyword in token_freq:
        if str(keyword) in str(words):
            token_freq[keyword] += 1

for words in df['Examination']:
    for keyword in token_freq:
        if str(keyword) in str(words):
            token_freq[keyword] += 1
```
